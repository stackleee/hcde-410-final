### Findings
#### Research Question 1: *How does the funding vary across the high schools within the Tacoma Public School District?*
To answer this question, I need to combine two CSV files containing different sets of data: one with the equity index and the other with block group IDs linked to each high school. The block group ID data was obtained from the Tacoma Open Data portal (https://data.cityoftacoma.org/datasets/8295bb884da94fe18a39e386a3cbd445_2/explore?location=47.224609%2C-122.420880%2C17.44). Since there was no specific data available on the exact funding of each school, I chose to use the equity index as a proxy for funding disparities. This index includes both economic and education-related factors, which will allow me to assess which schools are likely to be funded higher or lower based on these characteristics. By incorporating these factors into my analysis, I can make a more informed judgment about how funding disparities may affect different high schools in the Tacoma Public School District.

```python
import pandas as pd
import plotly.express as px

# Load the two CSV files into pandas DataFrames
equity_df = pd.read_csv('Equity_Index_2020_(Tacoma)_-_Data.csv')
highschool_df = pd.read_csv('TPS HighSchool BlockGroupID - Sheet1.csv')

# Strip any extra spaces in the column names to avoid mismatches
equity_df.columns = equity_df.columns.str.strip()
highschool_df.columns = highschool_df.columns.str.strip()

# Merge the two dataframes on the 'geoid' and 'Block Group ID' columns
merged_df = pd.merge(
    highschool_df[['Block Group ID', 'High School Name']],  # Include only relevant columns
    equity_df[['geoid', 'education_value']],
    left_on='Block Group ID',
    right_on='geoid',
    how='left'
)

# Group by 'High School Name' and calculate the average education_value for each school
bar_data = merged_df.groupby('High School Name')['education_value'].mean()

# Sort the data from lowest to highest education value
bar_data = bar_data.sort_values()

# Create a Plotly bar chart
fig = px.bar(
    bar_data,
    x=bar_data.index,
    y=bar_data.values,
    title='Average Education Equity by High School',
    labels={'x': 'High School Name', 'y': 'Education Equity Value'},
    text=bar_data.values,
    color=bar_data.values,
    color_continuous_scale='Blues'  # Apply a color scale
)

# Enhance the chart with customizations
fig.update_traces(
    texttemplate='%{text:.2f}', 
    textposition='outside',
    marker_line_width=1.5, 
    marker_line_color='black'
)
fig.update_layout(
    xaxis_title='High School Name',
    yaxis_title='Education Equity Value',
    xaxis=dict(tickangle=45),  # Rotate x-axis labels for readability
    title_font=dict(size=20, family='Arial, bold'),
    font=dict(size=14, family='Arial'),
    plot_bgcolor='rgba(0,0,0,0)',  # Transparent background
    paper_bgcolor='rgba(0,0,0,0)',  # Transparent container
    coloraxis_showscale=False,  # Hide the color scale
    margin=dict(l=50, r=50, t=50, b=100)  # Adjust margins for better spacing
)
fig.update_layout(
    xaxis_title='High School Name',
    yaxis_title='Education Equity Value',
    xaxis=dict(
        tickangle=45,  # Rotate x-axis labels for readability
        showgrid=False  # Disable gridlines for the x-axis
    ),
    yaxis=dict(
        showgrid=True,  # Enable gridlines for the y-axis
        gridcolor='lightgray',  # Set gridline color to light gray
        gridwidth=1,  # Set gridline thickness
        zeroline=True,  # Enable zero gridline
        zerolinecolor='black',  # Set the color of the zero gridline
        zerolinewidth=0.5  # Set the thickness of the zero gridline
    ),
    title_font=dict(size=20, family='Arial, bold'),
    font=dict(size=14, family='Arial'),
    plot_bgcolor='rgba(255,255,255,1)',  # White background for clarity
    paper_bgcolor='rgba(0,0,0,0)',  # Transparent container
    coloraxis_showscale=False,  # Hide the color scale
    margin=dict(l=50, r=50, t=50, b=100)  # Adjust margins for better spacing
)
fig.update_layout(
    xaxis_title='High School Name',
    yaxis_title='Education Equity Value',
    xaxis=dict(
        tickangle=45,  # Rotate x-axis labels for readability
        showgrid=False  # Disable gridlines for the x-axis
    ),
    yaxis=dict(
        showgrid=True,  # Enable gridlines for the y-axis
        gridcolor='lightgray',  # Set gridline color to light gray
        gridwidth=0.5  # Set gridline thickness
    ),
    title_font=dict(size=20, family='Arial, bold'),
    font=dict(size=14, family='Arial'),
    plot_bgcolor='rgba(255,255,255,1)',  # White background for clarity
    paper_bgcolor='rgba(0,0,0,0)',  # Transparent container
    coloraxis_showscale=False,  # Hide the color scale
    margin=dict(l=50, r=50, t=50, b=100)  # Adjust margins for better spacing
)


# Display the bar chart
fig.show()

# Save the sorted merged data to a new CSV file
merged_df.to_csv('Sorted_Merged_Schools_Education_Equity.csv', index=False)
```

```python
import pandas as pd
import plotly.express as px

# Load the provided CSV files
correct_funding_data_path = "Sorted_Merged_Schools_Education_Equity.csv"
correct_student_outcomes_path = "TP Student Outcomes - Sheet1 (1).csv"

# Read the data
funding_data = pd.read_csv(correct_funding_data_path)
student_outcomes = pd.read_csv(correct_student_outcomes_path)

# Clean and prepare the data
for col in [
    "Percentage of Students Below State Standards in ELA (Assessment SBA)",
    "Percentage of Students Below State Standards in Math (Assessment SBA)",
    "Graduation Rates",
    "Acceptance to Next Institution"
]:
    student_outcomes[col] = student_outcomes[col].str.replace('%', '', regex=False).astype(float)

# Merge datasets on the "High School Name" column
merged_data = pd.merge(
    funding_data, 
    student_outcomes, 
    left_on="High School Name", 
    right_on="High School Name"
)

# Function to plot relationships
def plot_relationship(y_column, y_label, title):
    fig = px.scatter(
        merged_data,
        x="education_value",
        y=y_column,
        title=title,
        labels={
            "education_value": "Education Value (Proxy for Funding)",
            y_column: y_label
        },
        hover_name="High School Name",
        trendline="ols"  # Ordinary Least Squares regression trendline
    )
    fig.update_layout(
        xaxis_title="Education Value (Proxy for Funding)",
        yaxis_title=y_label,
        template="plotly_white",
    )
    fig.show()

# Plot each outcome
plot_relationship(
    "Percentage of Students Below State Standards in Math (Assessment SBA)",
    "Percentage Below Standards in Math (%)",
    "Relationship Between Education Value and Percentage of Students Below Standards in Math"
)

plot_relationship(
    "Graduation Rates",
    "Graduation Rates (%)",
    "Relationship Between Education Value and Graduation Rates"
)

plot_relationship(
    "Acceptance to Next Institution",
    "Acceptance to Next Institution (%)",
    "Relationship Between Education Value and Acceptance to Next Institution"
)
```

```python
import pandas as pd
import plotly.express as px

# Load the provided CSV files
correct_funding_data_path = "Sorted_Merged_Schools_Education_Equity.csv"
correct_student_outcomes_path = "TP Student Outcomes - Sheet1 (1).csv"

# Read the data
funding_data = pd.read_csv(correct_funding_data_path)
student_outcomes = pd.read_csv(correct_student_outcomes_path)

# Clean and prepare the data
for col in [
    "Percentage of Students Below State Standards in ELA (Assessment SBA)",
    "Percentage of Students Below State Standards in Math (Assessment SBA)",
    "Graduation Rates",
    "Acceptance to Next Institution"
]:
    student_outcomes[col] = student_outcomes[col].str.replace('%', '', regex=False).astype(float)

# Merge datasets on the "High School Name" column
merged_data = pd.merge(
    funding_data, 
    student_outcomes, 
    left_on="High School Name", 
    right_on="High School Name"
)

# Sort the data by education value (funding) in ascending order
merged_data_sorted = merged_data.sort_values(by="education_value", ascending=True)

# Function to plot relationships with sorted data and color-coded points
def plot_relationship_sorted(y_column, y_label, title):
    fig = px.scatter(
        merged_data_sorted,
        x="education_value",
        y=y_column,
        color="High School Name",  # Assign a unique color for each high school
        title=title,
        labels={
            "education_value": "Education Value (Proxy for Funding)",
            y_column: y_label,
            "High School Name": "High School"
        },
        trendline="ols"  # Ordinary Least Squares regression trendline
    )
    fig.update_layout(
        xaxis_title="Education Value (Proxy for Funding)",
        yaxis_title=y_label,
        template="plotly_white",
        legend_title="High Schools"  # Label the legend
    )
    fig.show()

# Plot each outcome
plot_relationship_sorted(
    "Percentage of Students Below State Standards in Math (Assessment SBA)",
    "Percentage Below Standards in Math (%)",
    "Relationship Between Education Value and Percentage of Students Below Standards in Math"
)

plot_relationship_sorted(
    "Graduation Rates",
    "Graduation Rates (%)",
    "Relationship Between Education Value and Graduation Rates"
)

plot_relationship_sorted(
    "Acceptance to Next Institution",
    "Acceptance to Next Institution (%)",
    "Relationship Between Education Value and Acceptance to Next Institution"
)
```

```python
import pandas as pd
import plotly.express as px

# Load the provided CSV files
correct_funding_data_path = "Sorted_Merged_Schools_Education_Equity.csv"
correct_student_outcomes_path = "TP Student Outcomes - Sheet1 (1).csv"

# Read the data
funding_data = pd.read_csv(correct_funding_data_path)
student_outcomes = pd.read_csv(correct_student_outcomes_path)

# Clean and prepare the data
for col in [
    "Percentage of Students Below State Standards in ELA (Assessment SBA)",
    "Percentage of Students Below State Standards in Math (Assessment SBA)",
    "Graduation Rates",
    "Acceptance to Next Institution"
]:
    student_outcomes[col] = student_outcomes[col].str.replace('%', '', regex=False).astype(float)

# Merge datasets on the "High School Name" column
merged_data = pd.merge(
    funding_data, 
    student_outcomes, 
    left_on="High School Name", 
    right_on="High School Name"
)

# Sort the data by education value (funding) in ascending order
merged_data_sorted = merged_data.sort_values(by="education_value", ascending=True)

# Function to plot relationships with sorted data, color-coded points, and enhanced trendline
def plot_relationship_sorted(y_column, y_label, title):
    fig = px.scatter(
        merged_data_sorted,
        x="education_value",
        y=y_column,
        color="High School Name",  # Assign a unique color for each high school
        title=title,
        labels={
            "education_value": "Education Value (Proxy for Funding)",
            y_column: y_label,
            "High School Name": "High School"
        },
        trendline="ols"  # Ordinary Least Squares regression trendline
    )
    # Customize the trendline for better visibility
    fig.update_traces(marker=dict(size=8), selector=dict(mode="markers"))
    fig.update_traces(
        line=dict(color="red", width=3),  # Customize the trendline color and width
        selector=dict(mode="lines")
    )
    fig.update_layout(
        xaxis_title="Education Value (Proxy for Funding)",
        yaxis_title=y_label,
        template="plotly_white",
        legend_title="High Schools"  # Label the legend
    )
    fig.show()

# Plot each outcome
plot_relationship_sorted(
    "Percentage of Students Below State Standards in Math (Assessment SBA)",
    "Percentage Below Standards in Math (%)",
    "Relationship Between Education Value and Percentage of Students Below Standards in Math"
)

plot_relationship_sorted(
    "Graduation Rates",
    "Graduation Rates (%)",
    "Relationship Between Education Value and Graduation Rates"
)

plot_relationship_sorted(
    "Acceptance to Next Institution",
    "Acceptance to Next Institution (%)",
    "Relationship Between Education Value and Acceptance to Next Institution"
)
```
